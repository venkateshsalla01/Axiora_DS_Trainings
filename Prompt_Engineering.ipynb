# **Prompt Engineering**

Prompt Engineering is the art and science of designing effective instructions (called prompts ) to communicate with AI systems like ChatGPT, Gemini, Claude, LLaMA, or image generators like DALL·E and Midjourney.

A prompt = the input message that guides the AI to produce the desired output.


Prompt Engineering = creating prompts that are: 

• Clear 

• Specific

• Structured

• context-rich

• goal-oriented

# **Prompt Engineering Techniques**

There are 3 Prompting Techniques

**Zero Prompting**

Zero-shot prompting involves providing the model with a direct instruction or question without any additional context or examples. 

An example of this is idea generation, where the model is prompted to generate creative ideas or brainstorming solutions. Another example is summarization, or translation, where the model is asked to summarize or translate some piece of content.

**Example prompt:**  Sun rises in the east.

**Response:** Yes, that's correct

**One-Shot Prompting**

This method involves providing the model with one or more examples of the desired input-output pairs before presenting the actual prompt. This can help the model better understand the task and generate more accurate responses.

prompt --> Response

**Few shot prompting**

In Few shot prompting, you provide multiple examples before asking the actual questions.

Example:

Prompt1 --> Response 1

Prompt2 --> Response 2

Prompt3 --> Response 3

Then you can ask

Prompt4 --> Response 4

#### **Chain of Thought Prompting (COT)**

CoT prompting encourages the model to break down complex reasoning into a series of intermediate steps, leading to a more comprehensive and well-structured final output.

Example - Logical Reasoning

Right side of V -> B

Right side of T -> Y

Right side of X -> C

Right side of Y -> U

so:
    V T X Y -> B Y C U

what does V A N S mean?

To solve this logically

step by step reasoning:

Right of V -> B

Right of A -> S

Right of M -> (Comma)

Right of S -> D

**Zero shot COT Prompting**

Combines chain of thought prompting with zero-shot prompting by asking the model to perform reasoning steps, which may often produce better output.

You provide one example, it will generate step-by-step reasoning, then ask next question


It teaches the model how to think step-by-step

Instead of :

Question --> Answer

It becames:

### **Self consistency Prompting**

Self-consistency prompting is an advanced technique that improves Large Language Model (LLM) reliability by sampling multiple, diverse reasoning paths (using Chain-of-Thought (CoT)) and selecting the most consistent final answer via majority vote. It boosts performance in arithmetic and logic tasks by overcoming the limitations of greedy decoding.

**How self-consistency works**

Question --> one reasoning --> One Answer


Self consistency answer:

Question --> Multiple reasoning paths --> Different answers --> Majority voting --> Final Answer


### **Out-of-Date Learning**

Out-of-date learning in Large Language Models (LLMs) refers to the limitation where a model’s knowledge is frozen in time based on the training data cutoff date. Because LLMs are trained on vast datasets at specific points in time, they lack inherent knowledge of events, discoveries, API changes, or developments that occurred after their training was completed.

### **Role Playing prompt**

Role-playing prompting (or role-based prompting) is a technique in AI engineering where you assign a specific persona, character, or expertise level to the AI to improve the relevance, tone, and accuracy of its responses. By starting a request with phrases like "Act as..." or "You are...", you guide the AI to adopt a particular perspective.

**1. Role Assignment**

Tell the LLM what role it should take

Examples:
    You are an experienced software engineer
    
    You are a professional script writer

    You are a rocket scientist

**2. Actual Question**

Ask the main question or task

Example:
    write a code to generate a QR code for a website

**3.Output Instruction**

Tell the model how you want the output

**Examples:**  
    Json, CSV format, Python, Java, Yes/No, Detailed explanation etc..

**Example without Role**

Write a code to generate a QR code for a website.

**Example with Role playing**

You are an experienced software engineer.

Write a code to generate a QR code for a website.

Provide the code in python with step-by-step explanation.
