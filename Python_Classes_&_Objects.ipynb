### **Classes & Objects**

1. A Class is an object constructor or a "blueprint" for creating objects.

2. Objects are nothing but an encapsulation of variables and functions into a single entity.

3. Objects get their variables and functions from classes.

4. To create a class we use the keyword class.

5. The first string inside the class is called docstring which gives the brief description about theclass.

6. All classes have a function called __init__() which is always executed when the class isbeing initiated.

7. We can use __init__() function to assign values to object properties or other operationsthat are necessary to perform when the object is being created

8. The self parameter is a reference to the current instance of the class and is used to access class variables.

9. self must be the first parameter of any function in the class

10. The super() builtin function returns a temporary object of the superclass that allows us toaccess methods of the base class.

11. super() allows us to avoid using the base class name explicitly and to enable multipleinheritance.

**syntax**

```python
# Create a class with property "var1"
class MyClass:
    var1 = 10

# Create an object of class "MyClass"
obj1 = MyClass()

# Access the property
print(obj1.var1)
```

```python
# Create an employee class
class Employee:
    
    # __init__() function is used to assign values
    def __init__(self, name, empid):
        self.name = name
        self.empid = empid

    # Class Method
    def greet(self):
        print("Thanks for joining ABC Company {}!!".format(self.name))


# Create an employee object
emp1 = Employee("Saikumar", 34163)

print('Name :- ', emp1.name)
print('Employee ID :- ', emp1.empid)

emp1.greet()
```

```python
emp1.name='Basit'# Modify Object Properties
emp1.name
```

```python
del emp1.empid# Delete Object Properties
emp1.empid
```

```python
del emp1 # Delete the object
emp1
```

#### **Inheritance**

1. Inheritance is a powerful feature in object oriented programming.

2. Inheritance provides code reusability in the program because we can use an existing class(Super Class/ Parent Class / Base Class) to create a new class (Sub Class / Child Class /Derived Class) instead of creating it from scratch.

3. The child class inherits data definitions and methods from the parent class which facilitates thereuse of features already available. The child class can add few more definitions or redefine abase class method.

4. Inheritance comes into picture when a new class possesses the 'IS A' relationship with anexisting class. E.g Student is a person. Hence person is the base class and student is derivedclass.

```python
# Parent Class
class Person:

    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender

    def PersonInfo(self):
        print('Name :- {}'.format(self.name))
        print('Age :- {}'.format(self.age))
        print('Gender :- {}'.format(self.gender))


# Child Class
class Student(Person):

    def __init__(self, name, age, gender, studentid, fees):
        Person.__init__(self, name, age, gender)
        self.studentid = studentid
        self.fees = fees

    def StudentInfo(self):
        print('Student ID :- {}'.format(self.studentid))
        print('Fees :- {}'.format(self.fees))


# Child Class
class Teacher(Person):

    def __init__(self, name, age, gender, empid, salary):
        Person.__init__(self, name, age, gender)
        self.empid = empid
        self.salary = salary

    def TeacherInfo(self):
        print('Employee ID :- {}'.format(self.empid))
        print('Salary :- {}'.format(self.salary))


# Creating Student Object
stud1 = Student('Asif', 24, 'Male', 123, 1200)

print('Student Details')
print('---------------')

# Accessing Parent Class Method
stud1.PersonInfo()

# Accessing Child Class Method
stud1.StudentInfo()

print()


# Creating Teacher Object
teacher1 = Teacher('Basit', 36, 'Male', 456, 80000)

print('Employee Details')
print('---------------')

# Accessing Parent Class Method
teacher1.PersonInfo()

# Accessing Child Class Method
teacher1.TeacherInfo()
```

```python
# Parent Class
class Person:

    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender

    def PersonInfo(self):
        print('Name :- {}'.format(self.name))
        print('Age :- {}'.format(self.age))
        print('Gender :- {}'.format(self.gender))


# Child Class
class Student(Person):

    def __init__(self, name, age, gender, studentid, fees):
        Person.__init__(self, name, age, gender)
        self.studentid = studentid
        self.fees = fees

    def StudentInfo(self):
        print('Student ID :- {}'.format(self.studentid))
        print('Fees :- {}'.format(self.fees))


# Creating Student Object
stud1 = Student('Asif', 24, 'Male', 123, 1200)

print('Student Details')
print('---------------')

# PersonInfo() method present in Parent Class will be accessed
stud1.PersonInfo()

# StudentInfo() method present in Child Class
stud1.StudentInfo()

print()
```
