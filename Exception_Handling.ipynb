1. Python has many built-in exceptions (ArithmeticError, ZeroDivisionError, EOFError, IndexError,KeyError, SyntaxError, IndentationError, FileNotFoundError etc) that are raised when yourprogram encounters an error.

2. When the exception occurs Python interpreter stops the current process and passes it to thecalling process until it is handled. If exception is not handled the program will crash.

3. Exceptions in python can be handled using a try statement. The try block lets you test a blockof code for errors.

4. The block of code which can raise an exception is placed inside the try clause. The code thatwill handle the exceptions is written in the except clause.

5. The finally code block will execute regardless of the result of the try and except blocks.

6. We can also use the else keyword to define a block of code to be executed if no exceptions were raised.

7. Python also allows us to create our own exceptions that can be raised from the program using the raise keyword and caught using the
except clause. We can define what kind of error toraise, and the text to print to the user.


```python
import sys

try:
    print(100 / 0)
    # ZeroDivisionError will be encountered here. So the control will go to except

except:
    print(sys.exc_info()[1], 'Exception occurred')
    # This statement will be executed

else:
    print('No exception occurred')
    # This will be skipped as code block inside try raises exception

finally:
    print('Run this block of code always')
    # This will be always executed
```

```python
try:
    print(x)

except:
    print('Variable x is not defined')
```

```python
import os
try:
    os.remove("text1.txt") # FileNotFoundError will be encountered as "test1.txt is not available

except:
    print("BELOW EXCEPTION OCCURED")
    print(sys.exc_info()[1])

else:
    print('\nNo exception occured')

finally:
    print("\n Run this block of code always")
```

```python
import os

# Handling specific exceptions
try:
    x = int(input('Enter first number :- '))
    y = int(input('Enter second number :- '))
    
    # If input entered is non-zero then division works
    print(x / y)
    
    os.remove("test3.txt")

except NameError:
    print('NameError exception occurred')

except FileNotFoundError:
    print('FileNotFoundError exception occurred')

except ZeroDivisionError:
    print('ZeroDivisionError exception occurred')
```

```python
import os

# Handling specific exceptions
try:
    x = int(input('Enter first number :- '))
    y = int(input('Enter second number :- '))
    
    # If the input entered is zero then ZeroDivisionError occurs
    print(x / y)
    
    os.remove("test3.txt")

except NameError:
    print('NameError exception occurred')

except FileNotFoundError:
    print('FileNotFoundError exception occurred')

except ZeroDivisionError:
    print('ZeroDivisionError exception occurred')
```

```python
import sys

try:
    x = int(input('Enter first number :- '))
    
    if x > 50:
        # If value of x is greater than 50, raise ValueError exception
        raise ValueError(x)

except:
    print(sys.exc_info()[0])
```

**Built-in Exceptions**

```python
import sys

try:
    import math
    print(math.exp(1000))

except OverflowError:
    print(sys.exc_info())

else:
    print("Success, no error!")
```

```python
# AssertionError - This exception is raised when an assert statement fails
try:
    a = 50
    b = "Asif"
    
    assert a == b

except AssertionError:
    print("Assertion Exception Raised.")
```

```python
# ModuleNotFoundError - This exception is raised when an imported module does not exist
try:
    import MyModule

except ModuleNotFoundError:
    print("ModuleNotFoundError Exception Raised.")
```

```python
# KeyError - This exception is raised when key does not exist in a dictionary
try:
    mydict = {
        1: 'Asif',
        2: 'Basit',
        3: 'Michael'
    }
    
    print(mydict[4])

except KeyError:
    print("KeyError Exception Raised.")
```

```python
# IndexError - This exception is raised when an index of a sequence does not exist
try:
    mylist = [1, 2, 3, 4, 5, 6]
    
    print(mylist[10])

except IndexError:
    print("IndexError Exception Raised.")
```

```python
# TypeError - This exception is raised when two different datatypes are combined
try:
    a = 50
    b = "Asif"
    c = a / b

except TypeError:
    print("TypeError Exception Raised.")
```

```python
# AttributeError - This exception is raised when attribute reference or assignment fails
try:
    a = 10
    b = a.upper()  # int object has no attribute 'upper'
    print(b)

except AttributeError:
    print("AttributeError Exception Raised.")
```
