### **Time Series Analysis and Forecasting**

To understand how data changes over time, Time Series Analysis and Forecasting are used, which help track past patterns and predict future values. It is widely used in finance, weather, sales and sensor data.

1. Focuses on data collected at regular time intervals

2. Helps identify trends, seasonality and sudden changes

3. Useful for planning, prediction and decision-making

4. Common methods include ARIMA, exponential smoothing and machine learning models

Time series are commonly visualised using a line plot with time on X-axis and observed values on Y-axis. This visualization helps identify trends, fluctuations and underlying patterns.

**Importance of Time Series Analysis**

1. Predicting Future Trends: Helps forecast outcomes like demand, revenue or stock prices.

2. Detecting Patterns and Anomalies: Identifies cycles, seasonal effects and unusual behaviors.

3. Risk Mitigation: Provides early warnings to prevent potential losses.

4. Strategic Planning: Supports budgeting, staffing, inventory and policy decisions.

5. Competitive Advantage: Enables faster response and optimized operations.

**Components of Time Series Data**

1. Trend: Long-term movement of data showing increase, decrease or stability can be linear or nonlinear.

2. Seasonality: Regular patterns repeating at fixed intervals like daily, monthly or yearly, often due to seasons or events.

3. Cyclical Variations: Long-term fluctuations without fixed periods, linked to economic or business cycles.

4. Irregularity (Noise): Random, unpredictable variations caused by events, errors or other unforeseen factors.

**When Should You Use Time Series Forecasting**

1. Historical data is consistent and reliable.

2. Temporal patterns or trends exist in the data.

3. Future behavior depends on past observations.

**Types of Time Series**

**1. Univariate vs Multivariate**

1. Univariate: A univariate time series records only one variable over time, making it simpler to analyze and model.

2. Multivariate: A multivariate time series tracks multiple related variables together to show how they influence each other over time.

**2. Continuous vs Discrete Time Series**

1. Continuous: A continuous time series is observed at every moment or at very high frequency like ECG or sensor signals.

2. Discrete: A discrete time series is recorded at fixed intervals such as hourly, daily or monthly and is the most common format.

**3. Stationary vs Non-Stationary**

1. Stationary: A stationary series has a constant mean, variance and pattern over time with no trend or seasonality.
2. Non-Stationary: A non-stationary series shows changing patterns like trends or seasonality and often needs transformation before modeling.

### **Preprocessing Time Series Data**

Time series preprocessing involves cleaning, transforming and preparing data for analysis or forecasting. The main aim is to improve data quality, remove noise and make the series suitable for modeling. Common tasks include:

1. Handling Missing Values: Fill or interpolate missing observations to maintain continuity.

2. Dealing with Outliers: Identify and correct extreme values that can distort analysis.

3. Stationarity and Transformation: Apply techniques like differencing, detrending or deseasonalizing to stabilize mean and variance over time.

4. Scaling and Normalization: Standardize data to improve model performance.

5. Variance Stabilization: Apply transformations to reduce variability and improve predictability.

**Time Series Preprocessing Techniques**

1. Stationarity: Keeps mean and variance constant over time.
2. Differencing: Removes trends by subtracting previous values.
3. Moving Average (MA): Smooths data using a fixed-window average.
4. Exponential Moving Average (EMA): Weighted average giving more importance to recent data.
5. Missing Value Imputation: Fills gaps using interpolation or statistical methods.
6. Outlier Detection and Removal: Identifies and corrects extreme values.
7. Scaling: Adjusts value range for comparability.
8. Normalization: Standardizes data to a common scale.

**Time Series Analysis and Decomposition**

Time Series Analysis and Decomposition is used to study sequential data over time, understand patterns and break the series into its core components i.e trend, seasonality and residuals.

1. Autocorrelation Analysis: Measures correlation between a series and its lagged values to detect patterns.
2. Partial Autocorrelation (PACF): Finds direct correlation with lagged values, controlling for intermediate lags.
3. Trend Analysis: Identifies long-term direction or movement of the series (linear, exponential or nonlinear).
4. Seasonality Analysis: Detects periodic patterns at fixed intervals like daily, weekly or yearly.
5. Decomposition: Separates series into trend, seasonal and residual components for easier analysis.
6. STL (Seasonal-Trend decomposition using Loess): Decomposes series into seasonal, trend and residual components for modeling.
7. Rolling Correlation: Computes correlation between series over a moving window to track changing relationships.

**Time Series Visualization**

1. **Line Plots:** Line plots show how a variable changes over time, helping you spot trends, jumps and fluctuations easily.

2. **Seasonal Plots:** Seasonal plots display repeating patterns across months, weeks or seasons to highlight periodic behavior.

3. **Histograms and Density Plots:** These plots show the distribution of values in a time series, making it easy to understand common ranges and spread.

4. **Autocorrelation and Partial Autocorrelation Plots:** ACF and PACF reveal how past values influence current values, helping choose suitable forecasting models.

5. **Spectral Analysis:** Spectral analysis identifies dominant frequency cycles in the data, useful for spotting repeating long-term patterns.

6. **Decomposition Plots:** Decomposition plots split the series into trend, seasonal and residual parts to clearly show the underlying structure.

**Time Series Forecasting Algorithms**

**Machine Learning Models**
1. Random Forest: Ensemble model for regression based time series forecasting.
2. Gradient Boosting (GBM): Boosted trees to capture complex patterns.
3. Generalized Additive Models (GAM): Combines linear and non-linear trends in series.

**Deep Learning Models**
1. RNN: Recurrent networks for sequential dependencies.
2. LSTM: Long Short-Term Memory networks handle long-term dependencies.
3. GRU: Gated Recurrent Units, a simpler alternative to LSTM.
4. Sequence-to-Sequence: Encoder decoder architecture for complex time series.
5. CNN-based Models: Capture local temporal patterns using convolutional layers.


**Probabilistic and Bayesian Models**
1. Gaussian Processes: Probabilistic modeling with uncertainty estimation.
2. State Space Models: Represent series as latent states evolving over time.
3. Dynamic Linear Models (DLM): Time-varying linear models for forecasting.
4. Hidden Markov Models (HMM): Models series with underlying hidden states.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create sample data
dates = pd.date_range(start='2020-01-01', periods=100)

sales = np.random.randint(100, 200, 100).cumsum()

df = pd.DataFrame({
    'Date': dates,
    'Sales': sales
})

df.set_index('Date', inplace=True)

print(df.head())
```

```python
plt.figure(figsize=(12,6))
plt.plot(df.index, df['Sales'])
plt.title("Sales Time Series")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()
```

```python
train = df[:80]
test = df[80:]
```

```python
window = 5

df['Moving_Avg'] = df['Sales'].rolling(window=window).mean()

plt.figure(figsize=(12,6))
plt.plot(df['Sales'], label='Actual')
plt.plot(df['Moving_Avg'], label='Moving Average')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Build model
model = ARIMA(train['Sales'], order=(1,1,1))

# Train model
model_fit = model.fit()

# Forecast
forecast = model_fit.forecast(steps=len(test))

# Plot
plt.figure(figsize=(12,6))
plt.plot(train.index, train['Sales'], label='Train')
plt.plot(test.index, test['Sales'], label='Test')
plt.plot(test.index, forecast, label='Forecast')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

mae = mean_absolute_error(test['Sales'], forecast)

rmse = np.sqrt(mean_squared_error(test['Sales'], forecast))

print("MAE:", mae)
print("RMSE:", rmse)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
from sklearn.metrics import mean_absolute_error, mean_squared_error
import pmdarima as pm
```

```python
periods = 120
time_index = pd.date_range("2015-01-01", periods=periods, freq="M")
trend = np.linspace(50, 120, periods)
season = 10 * np.sin(2 * np.pi * time_index.month / 12)
noise = np.random.normal(0, 5, periods)

ts = trend + season + noise
```

```python
data = pd.DataFrame({"Date": time_index, "Value": ts})
data.set_index("Date", inplace=True)
data.head()
```

```python
plt.figure(figsize=(12,5))
plt.plot(data.Value)
plt.title("Time Series Plot")
plt.xlabel("Time")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

```python
#Handling missing values
data.Value = data.Value.interpolate()
```

```python
# outlier detection and cleaning (z-score method)

z = np.abs((data.Value - data.Value.mean()) / data.Value.std())
data["Value_clean"] = np.where(z > 3, data.Value.median(), data.Value)
```

```python
decomp = seasonal_decompose(data.Value_clean.dropna(), model="additive", period=12)
decomp.plot()
plt.show()
```
